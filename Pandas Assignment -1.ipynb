#### 4 . Find the index positions of items common in  series A and series B using numpy.where() and pd.Index() methods?


```python
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])
```

```python
ser_u = pd.Series(np.union1d(ser1, ser2)) 
ser_i = pd.Series(np.intersect1d(ser1, ser2))  
ser_u[~ser_u.isin(ser_i)]
```

#### 6 . How to filter valid emails from a series data using following pattern to be matched?


```python
input_data = pd.Series(['books at amazom.com', 'james@egypt.com', 'jones@yahoo.in', 'bigdata@gmail.com'])
pattern ='[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'
```

```python
result = input_data.str.findall(pattern)
print("Valid Emails are: ")
print([i for i in result if len(i) > 0])
```
